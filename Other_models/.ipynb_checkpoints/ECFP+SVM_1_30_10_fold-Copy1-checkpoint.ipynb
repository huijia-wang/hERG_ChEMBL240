```python
import numpy as np
import pandas as pd
from rdkit import Chem
from rdkit.Chem.rdchem import RWMol
from rdkit.Chem import Draw
from rdkit.Chem import AllChem
from rdkit.Chem import rdmolops
from rdkit.Chem.Draw import IPythonConsole
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from sklearn.metrics import confusion_matrix
```

```python
df_rdm = pd.read_csv('/home/is/huijia-w/hERG/Data_from_Structure_base_ChEMBL240/Updated_data/No_Balanced/ChEMBL240_1_30_1295_2052.csv')
df_rdm
```

```python
nump = np.zeros((3347, 1024))
label = np.zeros((3347, 1))
for i,word in enumerate(list(df_rdm['Canonical_smiles'])):
    mol = Chem.MolFromSmiles(word) #mol
    fp = AllChem.GetMorganFingerprintAsBitVect(mol,2,nBits=1024)
    nump[i, :] = fp
    
    lab = list(df_rdm['ClassD'])[i]
    label[i,:] = lab
    
print(nump)
print(label)
```

```python
plt.figure(figsize = (10,8))
plt.imshow(nump)
```

*********************************Performance****************************

```python
from sklearn.model_selection import KFold
from sklearn import svm
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

k_folds = 10
# Define the K-fold Cross Validator
kfold = KFold(n_splits=k_folds, shuffle=False)

# Start print
print("------Start-----------------")
# K-fold Cross Validation model evaluation
for fold, (train_ids, test_ids) in enumerate(kfold.split(nump)):
    print(f'FOLD {fold}')
    
    clf = svm.SVC()# use RF to classify
    clf.fit(nump[train_ids], label[train_ids].reshape(-1,))
    
    pred = clf.predict(nump[test_ids])
    cl = classification_report(label[test_ids], pred, output_dict=True)
    print("FOLD:", fold, ", report: ", cl)
    
    cm1 = confusion_matrix(label[test_ids], pred)
    tn, fp, fn, tp = confusion_matrix(label[test_ids], pred).ravel()
    # print(cm1)
    print("tn, fp, fn, tp: ", tn, fp, fn, tp)
    from sklearn import metrics
    fpr, tpr, thresholds = metrics.roc_curve(label[test_ids], pred, pos_label=1)
    print("AUC: ", metrics.auc(fpr, tpr))
    GHa = (tp*((tp+fp)+(tp+fn))) / (2*(tp+fp)*(tp+fn))
    GHi = (tn*((tn+fn)+(tn+fp))) / (2*(tn+fn)*(tn+fp))
    print("GHa, GHi: ",GHa, GHi)
    NPV = tn/(tn+fn)
    print("NPV",NPV)
    print("-----------------------------")
    
```

```python
col = []
for i in range(10):
    name = "Acc_"+str(i+1)
    col.append(name)
for i in range(10):
    name = "Prs_"+str(i+1)
    col.append(name)
for i in range(10):
    name = "Rec_"+str(i+1)
    col.append(name)
for i in range(10):
    name = "F1_"+str(i+1)
    col.append(name)
for i in range(10):
    name = "AUC_"+str(i+1)
    col.append(name)
for i in range(10):
    name = "GHa_"+str(i+1)
    col.append(name)
for i in range(10):
    name = "GHi_"+str(i+1)
    col.append(name)
for i in range(10):
    name = "NPV_"+str(i+1)
    col.append(name)
    
import pandas as pd
datfrm = pd.DataFrame(columns=col, index=['1_1_class1', '1_1_class2'])
# datfrm['Acc_1'][0]
from sklearn.model_selection import KFold
from sklearn import svm
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

datfrm = pd.DataFrame(columns=col, index=['1_1_class1', '1_1_class2'])

k_folds = 10
# Define the K-fold Cross Validator
kfold = KFold(n_splits=k_folds, shuffle=False)

Pre = [[],[],[],[],[],[],[],[],[],[]]
Lab = [[],[],[],[],[],[],[],[],[],[]]

# Start print
print("------Start-----------------")
# K-fold Cross Validation model evaluation
for fold, (train_ids, test_ids) in enumerate(kfold.split(nump)):
    print(f'FOLD {fold}')
    
    clf = svm.SVC()# use RF to classify
    clf.fit(nump[train_ids], label[train_ids].reshape(-1,))
    
    pred = clf.predict(nump[test_ids])
    Pre[fold].append(pred)
    Lab[fold].append(label[test_ids])
    
    cls_rpt = classification_report(label[test_ids], pred, output_dict=True)
#     print(cls_rpt)
    acc_name = "Acc_" + str(fold+1)
    datfrm[acc_name][0] = cls_rpt["accuracy"]
#     print(prs_name)
    prs_name = "Prs_" + str(fold+1)
    datfrm[prs_name][0] = cls_rpt["0.0"]["precision"]
    datfrm[prs_name][1] = cls_rpt["1.0"]["precision"]
    
    rec_name = "Rec_"+ str(fold+1)
    datfrm[rec_name][0] = cls_rpt["macro avg"]["recall"]
    
    f1_name = "F1_"+ str(fold+1)
    datfrm[f1_name][0] = cls_rpt["macro avg"]["f1-score"]
    
    
    print("fold: ", fold, "\n", classification_report(Lab[fold][0], Pre[fold][0], output_dict=True))
    cm1 = confusion_matrix(Lab[fold][0], Pre[fold][0])
    tn, fp, fn, tp = confusion_matrix(Lab[fold][0], Pre[fold][0]).ravel()
    print("fold", fold,"\n")
    print(cm1)
    print("tn, fp, fn, tp: ", tn, fp, fn, tp)
    fpr, tpr, thresholds = metrics.roc_curve(Lab[fold][0], Pre[fold][0], pos_label=1)
    print("AUC: ", metrics.auc(fpr, tpr))
    
    AUC_name = "AUC_"+ str(fold+1)
    datfrm[AUC_name][0] = metrics.auc(fpr, tpr)
    
    
    GHa = (tp*((tp+fp)+(tp+fn))) / (2*(tp+fp)*(tp+fn))
    GHi = (tn*((tn+fn)+(tn+fp))) / (2*(tn+fn)*(tn+fp))
    print("GHa, GHi: ",GHa, GHi)
    
    GHa_name = "GHa_"+ str(fold+1)
    datfrm[GHa_name][0] = GHa
    GHi_name = "GHi_"+ str(fold+1)
    datfrm[GHi_name][0] = GHi
    
    NPV = tn/(tn+fn)
    print("NPV",NPV)
    NPV_name = "NPV_"+ str(fold+1)
    datfrm[NPV_name][0] = NPV
    
    print("===================")
# #     
#     print("FOLD:", fold, ", report: ", cl)
    
#     cm1 = confusion_matrix(label[test_ids], pred)
#     tn, fp, fn, tp = confusion_matrix(label[test_ids], pred).ravel()
#     # print(cm1)
#     print("tn, fp, fn, tp: ", tn, fp, fn, tp)
#     from sklearn import metrics
#     fpr, tpr, thresholds = metrics.roc_curve(label[test_ids], pred, pos_label=1)
#     print("AUC: ", metrics.auc(fpr, tpr))
#     GHa = (tp*((tp+fp)+(tp+fn))) / (2*(tp+fp)*(tp+fn))
#     GHi = (tn*((tn+fn)+(tn+fp))) / (2*(tn+fn)*(tn+fp))
#     print("GHa, GHi: ",GHa, GHi)
#     NPV = tn/(tn+fn)
#     print("NPV",NPV)
#     print("-----------------------------")
    
```
